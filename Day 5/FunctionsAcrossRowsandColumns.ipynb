```python
import numpy as np
matrix = np.array([[10,20,30],
                   [40,50,60],
                   [70,80,90]])

print("Matrix:")
print(matrix)

print("Row Sum:", np.sum(matrix, axis=1))
print("Column Sum:", np.sum(matrix, axis=0))

print("Row Mean:", np.mean(matrix, axis=1))
print("Column Mean:", np.mean(matrix, axis=0))
```
