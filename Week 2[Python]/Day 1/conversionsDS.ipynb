```python
numbers_list = [1, 2, 3, 2, 4]
numbers_set = set(numbers_list)
print("List:", numbers_list)
print("List to Set:", numbers_set)

set_to_list = list(numbers_set)
print("Set to List:", set_to_list)

tuple_data = (5, 6, 7)
tuple_to_list = list(tuple_data)
print("Tuple:", tuple_data)
print("Tuple to List:", tuple_to_list)

list_data = [8, 9, 10]
list_to_tuple = tuple(list_data)
print("List:", list_data)
print("List to Tuple:", list_to_tuple)
```
